```python
import pandas as pd
from pyprojroot import here
```

```python
data = pd.read_csv(here() / 'data/base/data.csv')
```

```python
def a(x):
    """
    """
    result = pd.DataFrame({
        col: x[col].value_counts().reindex()
            for col in x.columns
    })

    result = result.fillna(0) / result.sum()

    return result
```

```python
table = data.iloc[:,30:60].groupby([data['Country'], data['Hospital Name']]).apply(a)
```

```python
b = table.groupby(level=[0,2]).sum()
```

```python
b = table.groupby(level=[0,2]).sum()
```

```python
c = b / b.groupby(level='Country').transform('sum')
```
