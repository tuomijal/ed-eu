```python
import pandas as pd
from pyprojroot import here
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(here() / 'data/base/data.csv')
```

```python
a = data.iloc[:,[74, 63]]
```

```python
a.columns = ['freq', 'severity']
```

```python
order = ['Not at all', 'Slight', 'Moderate', 'Severe', 'Very severe']
```

```python
a['severity'] = pd.Categorical(a['severity'], categories=order, ordered=True)
```

```python
a.severity
```

```python
fig, ax = plt.subplots()

a.plot.box(by='severity', ax=ax)
#ax.set_xticklabels(order)
```
