```python
import pandas as pd
from pyprojroot import here

import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(here() / 'data/base/data.csv')
```

```python
b = data.iloc[:,75].value_counts()
```

```python
fig, ax = plt.subplots(figsize=(3,3))

# Create the pie (donut) chart
wedges, texts, autotexts = ax.pie(
    b.values,
    labels=['No', 'Yes'],
    autopct='%1.0f%%',
    startangle=90,
    colors=['.9', '#2F3088'],
    wedgeprops={'width': 0.4},
    pctdistance=.4
)

# ax.set_title(
#     'Do you currently use a patient volume forecasting or \n' \
#     'crowding early warning software to guide decision making',
#     fontsize=10
#     )

plt.savefig(here() / 'output/plots/usage.png', 
            bbox_inches='tight', 
            dpi=300)
```
