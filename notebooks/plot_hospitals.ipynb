```python
import pandas as pd
from pyprojroot import here
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(here() / 'data/base/data.csv')
```

```python
counts = data['Hospital Name'].value_counts().sort_values(ascending=False)
```

```python
fig, ax = plt.subplots(figsize=(4,10))

counts[counts > 1].plot(kind='barh', ax=ax)
```
