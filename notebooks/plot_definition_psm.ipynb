```python
import pandas as pd
from pyprojroot import here
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(here() / 'data/base/data.csv')
```

```python
a = data.iloc[:,[22, 65]].groupby('Country').value_counts().unstack()
```

```python
a.fillna(0)
```

```python
a.div(a.sum(axis=1), axis=0).fillna(0).round(2)
```

```python
values = data.iloc[:,65].value_counts()
```

```python
values = (values / values.sum()) * 100
values = values.round(0)
```

```python
fig, ax = plt.subplots(figsize=(6,2))

values.plot.barh(
    ax=ax,
    color='lightgrey',
    width=.8
    )

ax.set_ylabel('')
ax.set_xlabel('Proportion of responses (%)')
ax.set_xlim([0,100])

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.savefig(here() / 'output/plots/definition.png', dpi=300, bbox_inches='tight')
```
